### Step. Import the necessary libraries

```python
import pandas as pd
```

### Step. Import the dataset from this address.
### Step. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo
```

### Step. What is the price of each item?
print a data frame with only two columns item_name and item_price

```python
chipo[chipo['quantity']==1][['item_name', 'item_price']]
```

### Step. Sort by the name of the item

```python
chipo.sort_values(by=['item_name'])
```

### Step. What was the quantity of the most expensive item ordered?

```python
def slicing(string):
    return string[1:]

chipo['item_price_float']= chipo['item_price'].apply(slicing).astype('float')
chipo.sort_values(by=['item_price_float'], ascending=False).head(1)
```

```python
prices = [ float(value[1 : -1]) for value in chipo.item_price ]
chipo.item_price = prices
```
